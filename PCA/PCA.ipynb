```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("US_Accidents_June20_CLEANED4.csv")
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        le = LabelEncoder()
        df[col].fillna("None", inplace=True)
        le.fit(list(df[col].astype(str).values))
        df[col] = le.transform(list(df[col].astype(str).values))
    else:
        df[col].fillna(-999, inplace=True)
```

```python
numericData_lst=['Temperature(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Speed(mph)','newSeverity','Hour']
df=df[numericData_lst].copy()
df.head()
```

```python
df.shape
```

```python
# Separating out the features
#x = df.loc[:, features].values
# Separating out the target
#y = df.loc[:,['newSeverity']].values
# Standardizing the features
df = df[df.columns.difference(['Unnamed: 0','iforest','newSeverity'])]
x = StandardScaler().fit_transform(df)
```

```python
x
```

```python
from sklearn.decomposition import PCA
numComp = 28
pca = PCA(n_components=numComp)
principalComponents = pca.fit_transform(x)
```

```python
columns = []
for i in range(1,numComp+1):
    columns.append("PC" + str(i))
columns
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = columns)
```

```python
principalDf
```

```python
autovalores = pca.explained_variance_ratio_
autovalores
```
